```python
# Separate the Features and target variable

X = Data.drop('Mental_LossYear',axis=1)

Y = Data['Mental_LossYear']

X.head(10)

```

```python
Y.head(10)
```

```python
# Scaling of data

X.head(10)


from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
sc = MinMaxScaler()

x = X

for i in x.columns:
  print(i);
  if(i == 'Country' or i == 'Year'):
    continue
  x[i] = sc.fit_transform(x[[i]])


X = x
```

```python
# Now Do the encoding

enc  = LabelEncoder()

x = X
for i in x.columns:
  if( x[i].dtype != 'object'):
    continue
  else:
    x[i] = enc.fit_transform(x[[i]])

X = x

```

```python
plt.figure(figsize=(10,8))
sns.heatmap(X.corr(),cmap='Blues',annot=True)
```

```python
# train_split the Data

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
# Model building

from sklearn.linear_model  import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
print(y_pred)

from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
print("MSE",mean_squared_error(y_test,y_pred))

print("Root_Mean_Squared_Error",np.sqrt(mean_squared_error(y_test,y_pred)))
print('MAE(MEAN Absolute Error)', mean_absolute_error(y_test,y_pred))
print("R2 Score--" , r2_score(y_test,y_pred))

```

```python
# Model 2 --- using the Random forese regressior

from sklearn.ensemble import RandomForestRegressor
model2 = RandomForestRegressor(n_estimators=100,random_state=0)

model2.fit(x_train,y_train)
```

```python
y_pred2 = model2.predict(x_test)
print(y_pred2)

print(r2_score(y_test,y_pred2))

print(mean_absolute_error(y_test,y_pred2))

print(mean_squared_error(y_test,y_pred2))
```

```python
print("Best model Score Which is Random Foreset Regerssor : ",r2_score(y_test,y_pred2) * 100)
```
